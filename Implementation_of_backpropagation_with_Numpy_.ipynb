```python
import numpy as np

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))   # FIXED: added minus sign

def sigmoid_derivative(x):
    return x * (1 - x)

# Training data (XOR)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [1],
              [1],
              [0]])

# Initialize weights
np.random.seed(42)

input_neurons = 2
hidden_neurons = 2
output_neurons = 1

W1 = np.random.randn(input_neurons, hidden_neurons)
b1 = np.zeros((1, hidden_neurons))

W2 = np.random.randn(hidden_neurons, output_neurons)
b2 = np.zeros((1, output_neurons))

learning_rate = 0.1
epochs = 10000

# Training
for epoch in range(epochs):

    # Forward propagation
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)

    # Loss (MSE)
    loss = np.mean((y - y_pred) ** 2)

    # Backpropagation
    d_output = (y - y_pred) * sigmoid_derivative(y_pred)

    d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(a1)

    # Update weights
    W2 += np.dot(a1.T, d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += np.dot(X.T, d_hidden) * learning_rate   # FIXED typo
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    # Print loss
    if epoch % 2000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Testing
print("\nFinal Predictions:")
print(y_pred.round(3))
```
